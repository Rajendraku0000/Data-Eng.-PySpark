```
from pyspark.sql.functions import lit,col,explode
import pyspark.sql.functions as f
```

```
df2=df.withColumn("splitdata",f.split("value"," "))
display(df2)
```

```
df3=df2.withColumn("splitedData",explode("splitdata")).select("splitedData")
display(df3)
```

```
word_count=df3.groupBy("splitedData").count().orderBy(col("count"),ascending=False)
```
